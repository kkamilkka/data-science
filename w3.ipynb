```python
import pandas as pd #import pandas
```

creating a table from the given wikipedia link

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M' 
```

```python
df = pd.read_html(url)
df = df[0]
df
```

removal 'not assigned' row from 'boruogh' column 

```python
indexNames = df[ (df['Borough'] == 'Not assigned')].index
df.drop(indexNames , inplace=True)
df.head(50)
```

replacement 'not assigned' rows in Neighbourhood column 

```python
df.loc[df['Neighbourhood']=='Not assigned', ['Neighbourhood']] = df['Borough']
df.head(50)
```

```python
df = df.reset_index(drop=True)
df
```

grouping in a 'Neighbourhood' column

```python
df1 = df.groupby('Postcode')['Neighbourhood'].apply(', '.join).reset_index()
df1.head()
```

```python
df = pd.merge( df[['Postcode','Borough']], df1[['Postcode','Neighbourhood']], on='Postcode')
df.head(50) 
```

```python
import geocoder

df2 = df.copy()

postal_codes = df2['Postcode'].tolist()

# Search for the Latitude and Longitude of each postal code
for postal_code in postal_codes:
    # initialize your variable to None
    lat_lng_coords = None

    # loop until you get the coordinates
    print('{}, Toronto, Ontario'.format(postal_code))
    while(lat_lng_coords is None):
      g = geocoder.google('{}, Toronto, Ontario'.format(postal_code))

      lat_lng_coords = g.latlng

    latitude = lat_lng_coords[0]
    longitude = lat_lng_coords[1]
    
    df2.loc[df2.Postcode == postal_code, 'Latitude'] = latitude
    df2.loc[df2.Postcode == postal_code, 'Longitude'] = longitude
```
